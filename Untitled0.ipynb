```python
import os
import zipfile
import csv
from PIL import Image, ExifTags
from pillow_heif import register_heif_opener
import piexif
import xlsxwriter

# Register HEIC support for Pillow
register_heif_opener()

# Function to extract EXIF metadata
def get_exif(filename):
    try:
        image = Image.open(filename)
        # For JPEG and similar formats
        if hasattr(image, "_getexif"):
            return image._getexif()
        # For HEIC files
        elif filename.lower().endswith(".heic"):
            exif_data = image.info.get("exif", None)
            if exif_data:
                return piexif.load(exif_data)
            return None
        else:
            print(f"No EXIF support for file: {filename}")
            return None
    except Exception as e:
        print(f"Error extracting EXIF from {filename}: {e}")
        return None

# Function to extract GPS metadata
def get_geotagging(exif):
    if not exif:
        return None
    # GPS IFD tag
    gps_tag = piexif.ExifIFD.MakerNote
    gps_info = exif.get("GPS", {})
    if not gps_info:
        return None
    return gps_info

# Convert GPS coordinates to DMS format with direction
def convert_to_dms_with_direction(value, ref):
    if not value or not ref:
        return None
    degrees = float(value[0][0]) / float(value[0][1])
    minutes = float(value[1][0]) / float(value[1][1])
    seconds = float(value[2][0]) / float(value[2][1])
    direction = ref.strip()
    return f"{int(degrees)}º {int(minutes)}' {seconds:.2f}\" {direction}"

# Construct Google Maps location URL
def construct_google_maps_url(gps_info):
    if not gps_info:
        return None
    latitude = gps_info.get(piexif.GPSIFD.GPSLatitude)
    latitude_ref = gps_info.get(piexif.GPSIFD.GPSLatitudeRef)
    longitude = gps_info.get(piexif.GPSIFD.GPSLongitude)
    longitude_ref = gps_info.get(piexif.GPSIFD.GPSLongitudeRef)
    if latitude and longitude and latitude_ref and longitude_ref:
        # Convert latitude and longitude to decimal degrees
        lat = (latitude[0][0] / latitude[0][1] +
               latitude[1][0] / (60 * latitude[1][1]) +
               latitude[2][0] / (3600 * latitude[2][1])) * (1 if latitude_ref.decode() == "N" else -1)
        lon = (longitude[0][0] / longitude[0][1] +
               longitude[1][0] / (60 * longitude[1][1]) +
               longitude[2][0] / (3600 * longitude[2][1])) * (1 if longitude_ref.decode() == "E" else -1)
        return f"https://www.google.com/maps?q={lat},{lon}"
    return None

# Extract Created Date from EXIF metadata
def get_created_date(exif):
    if not exif:
        return None
    if "Exif" in exif:
        date_tag = piexif.ExifIFD.DateTimeOriginal
        return exif["Exif"].get(date_tag, b"").decode("utf-8") if date_tag in exif["Exif"] else None
    return None

# Specify ZIP file and extraction directory
zip_file_name = "/JioCloud-Download (1).zip"  # Replace with your ZIP file path
output_dir = "/content/extracted_images"
os.makedirs(output_dir, exist_ok=True)

# Extract ZIP file
with zipfile.ZipFile(zip_file_name, 'r') as zip_ref:
    zip_ref.extractall(output_dir)

# Prepare output Excel file
output_excel = "/content/metadata_output.xlsx"
owner_name = "Shivee"

# Write data to Excel file
workbook = xlsxwriter.Workbook(output_excel)
worksheet = workbook.add_worksheet()

# Add headers to the Excel sheet
headers = ["File Name", "Owner", "Created Date", "Latitude", "Longitude", "Location"]
for col_num, header in enumerate(headers):
    worksheet.write(0, col_num, header)

row = 1
for root, dirs, files in os.walk(output_dir):
    for file in files:
        file_path = os.path.join(root, file)
        try:
            # Get EXIF metadata
            exif_data = get_exif(file_path)
            created_date = get_created_date(exif_data) if exif_data else None
            gps_data = get_geotagging(exif_data) if exif_data else None
            latitude = convert_to_dms_with_direction(gps_data.get(piexif.GPSIFD.GPSLatitude), gps_data.get(piexif.GPSIFD.GPSLatitudeRef).decode()) if gps_data else None
            longitude = convert_to_dms_with_direction(gps_data.get(piexif.GPSIFD.GPSLongitude), gps_data.get(piexif.GPSIFD.GPSLongitudeRef).decode()) if gps_data else None
            location_url = construct_google_maps_url(gps_data) if gps_data else None

            # Write metadata to Excel
            worksheet.write(row, 0, file)
            worksheet.write(row, 1, owner_name)
            worksheet.write(row, 2, created_date)
            worksheet.write(row, 3, latitude)
            worksheet.write(row, 4, longitude)
            worksheet.write(row, 5, location_url)
            row += 1
        except Exception as e:
            print(f"File: {file} - Error: {str(e)}")

workbook.close()
print(f"Metadata extraction completed. Excel saved at: {output_excel}")
```
